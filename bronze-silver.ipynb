```
#fetching all the files into individual temp variables for further transformation

customer = spark.read.format("csv").option("header","true").load("/mnt/bronze/Customer.csv")
customerAddress = spark.read.format("csv").option("header", "true").load("/mnt/bronze/CustomerAddress.csv")
product = spark.read.format("csv").option("header", "true").load("/mnt/bronze/Product.csv")
productCategory = spark.read.format("csv").option("header","true").load("/mnt/bronze/ProductCategory.csv")
productDescription = spark.read.format("csv").option("header", "true").load("/mnt/bronze/ProductDescription.csv")
productModel = spark.read.format("csv").option("header","true").load("/mnt/bronze/ProductModel.csv")
productModelProductDescription = spark.read.format("csv").option("header","true").load("/mnt/bronze/ProductModelProductDescription.csv")
salesAddress = spark.read.format("csv").option("header","true").load("/mnt/bronze/SalesLT.Address.csv")
salesOrderDetail = spark.read.format("csv").option("header", "true").load("/mnt/bronze/SalesOrderDetail.csv")
salesOrderHeader = spark.read.format("csv").option("header","true").load("/mnt/bronze/SalesOrderHeader.csv")
```

```
customer.printSchema()
customerAddress.printSchema()
product.printSchema()
productCategory.printSchema()
productDescription.printSchema()
productModel.printSchema()
productModelProductDescription.printSchema()
salesAddress.printSchema()
salesOrderDetail.printSchema()
salesOrderHeader.printSchema()
```

```
#customer datatype changed
customer = customer.withColumn("CustomerID", col("CustomerID").cast(IntegerType())) \
    .withColumn("ModifiedDate", col("ModifiedDate").cast(DateType())) \
    .withColumn("Phone", col("Phone").cast(IntegerType()))
```

```
display(product)
```

```
customerAddress = customerAddress.withColumn("CustomerID", col("CustomerID").cast(IntegerType()))\
    .withColumn("ModifiedDate", col("ModifiedDate").cast(DateType()))

customerAddress.show()
```

```
product = product.withColumn("ProductID", col("ProductID").cast(IntegerType()))\
    .withColumn("ProductNumber", col("ProductNumber").cast(IntegerType()))\
        .withColumn("StandardCost", col("StandardCost").cast(FloatType()))\
            .withColumn("ListPrice", col("ListPrice").cast(FloatType()))\
                .withColumn("Weight", col("Weight").cast(FloatType()))\
                    .withColumn("ProductCategoryID", col("ProductCategoryID").cast(IntegerType()))\
                        .withColumn("ProductModelID", col("ProductModelID").cast(IntegerType()))\
                            .withColumn("SellStartDate", col("SellStartDate").cast(DateType()))\
                                .withColumn("SellEndDate", col("SellEndDate").cast(DateType()))\
                                    .withColumn("DiscontinuedDate", col("DiscontinuedDate").cast(DateType()))\
                                        .withColumn("ModifiedDate", col("ModifiedDate").cast(DateType()))
```

```
productCategory = productCategory.withColumn("ProductCategoryID", col("ProductCategoryID").cast(IntegerType()))\
    .withColumn("ParentProductCategoryID", col("ParentProductCategoryID").cast(IntegerType()))\
        .withColumn("ModifiedDate", col("ModifiedDate").cast(DateType()))
```

```
productDescription = productDescription.withColumn("ProductDescriptionID", col("ProductDescriptionID").cast(IntegerType()))\
    .withColumn("ModifiedDate", col("ModifiedDate").cast(DateType()))
```

```
productModel = productModel.withColumn("ProductModelID", col("ProductModelID").cast(IntegerType()))\
    .withColumn("ModifiedDate", col("ModifiedDate").cast(DateType()))
```

```
productModelProductDescription = productModelProductDescription.withColumn("ProductModelID", col("ProductModelID").cast(IntegerType()))\
    .withColumn("ProductDescriptionID", col("ProductDescriptionID").cast(IntegerType()))\
        .withColumn("ModifiedDate", col("ModifiedDate").cast(DateType()))
```

```
salesAddress = salesAddress.withColumn("AddressID", col("AddressID").cast(IntegerType()))\
    .withColumn("ModifiedDate", col("ModifiedDate").cast(DateType()))
    
```

```
salesOrderDetail = salesOrderDetail.withColumn("SalesOrderID", col("SalesOrderID").cast(IntegerType()))\
    .withColumn("SalesOrderDetailID", col("SalesOrderDetailID").cast(IntegerType()))\
        .withColumn("OrderQty", col("OrderQty").cast(IntegerType()))\
            .withColumn("ProductID", col("ProductID").cast(IntegerType()))\
                .withColumn("UnitPrice", col("UnitPrice").cast(FloatType()))\
                    .withColumn("UnitPriceDiscount", col("UnitPriceDiscount").cast(FloatType()))\
                        .withColumn("ModifiedDate", col("ModifiedDate").cast(DateType()))
```

```
salesOrderHeader = salesOrderHeader.withColumn("SalesOrderID", col("SalesOrderID").cast(IntegerType()))\
    .withColumn("RevisionNumber", col("RevisionNumber").cast(IntegerType()))\
        .withColumn("OrderDate", col("OrderDate").cast(DateType()))\
            .withColumn("DueDate", col("DueDate").cast(DateType()))\
                .withColumn("ShipDate", col("ShipDate").cast(DateType()))\
                        .withColumn("Status", col("Status").cast(IntegerType()))\
                            .withColumn("AccountNumber", col("AccountNumber").cast(IntegerType()))\
                                .withColumn("CustomerID", col("CustomerID").cast(IntegerType()))\
                                    .withColumn("ShipToAddressID", col("ShipToAddressID").cast(IntegerType()))\
                                        .withColumn("BillToAddressID", col("BillToAddressID").cast(IntegerType()))\
                                            .withColumn("SubTotal", col("SubTotal").cast(FloatType()))\
                                                .withColumn("TaxAmt", col("TaxAmt").cast(FloatType()))\
                                                    .withColumn("Freight", col("Freight").cast(FloatType()))\
                                                        .withColumn("TotalDue", col("TotalDue").cast(FloatType()))\
                                                            .withColumn("ModifiedDate", col("ModifiedDate").cast(DateType()))
```

```
dbutils.fs.ls("/mnt/silver")
```

```
#code to load the transformed data csv files to silver tier
customer.write.option("header","true").csv("/mnt/silver/Customer")
customerAddress.write.option("header","true").csv("/mnt/silver/CustomerAddress")
product.write.option("header","true").csv("/mnt/silver/Product")
productCategory.write.option("header","true").csv("/mnt/silver/ProductCategory")
productDescription.write.option("header","true").csv("/mnt/silver/ProductDescription")
productModel.write.option("header","true").csv("/mnt/silver/ProductModel")
productModelProductDescription.write.option("header","true").csv("/mnt/silver/ProductModelProductDescription")
salesAddress.write.option("header","true").csv("/mnt/silver/SalesAddress")
salesOrderDetail.write.option("header","true").csv("/mnt/silver/SalesOrderDetail")
salesOrderHeader.write.option("header","true").csv("/mnt/silver/SalesOrderHeader")
```
